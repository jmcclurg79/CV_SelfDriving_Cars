```python
# --- IMPORTS ---
from google.colab import drive
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# --- NLTK DOWNLOADS (Run once) ---
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    nltk.download('stopwords')
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    nltk.download('punkt')

# --- CONFIGURATION ---
# Define the corrected path after mounting (lowercase gdrive is critical)
# Ensure this path matches the location of your 'datasets' folder exactly!
BASE_PATH = '/content/gdrive/MyDrive/datasets'

file_2006 = '2006-GWBush.txt'
file_2005 = '2005-GWBush.txt'

ps = PorterStemmer()
english_stopwords = set(stopwords.words('english'))

# --- 1. FILE READING FUNCTION ---
def read_speech_file(filename):
    """Reads a file from the defined Google Drive path."""
    filepath = os.path.join(BASE_PATH, filename)
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            print(f"--- SUCCESS: Reading {filename} ---")
            return f.read()
    except FileNotFoundError:
        print(f"*** ERROR: File not found at {filepath}. Please check your path/spelling. ***")
        return None

# --- 2. NLP PROCESSING FUNCTION ---
def process_text(text, file_label):
    """Performs tokenization, filtering, sorting, and stemming."""
    print(f"\n--- PROCESSING: {file_label} ---")

    # a) Tokenize
    words = word_tokenize(text)

    # b) Remove Stop Words
    # Filter out words that are in the stop word set (and handle punctuation/numbers)
    filtered_words = [
        word for word in words
        # Check if the token is a word (alphabetic) and not a stopword
        if word.isalpha() and word.lower() not in english_stopwords
    ]

    # c) Sort (Creates a NEW, alphabetically sorted list)
    sorted_list = sorted(filtered_words)

    # d) Stem and Display (for verification)
    stemmed_results = [ps.stem(w) for w in sorted_list]

    print(f"Total tokens after filtering: {len(filtered_words)}")
    print("Top 20 Sorted & Stemmed Words:")
    # Print a sample of the results
    print(stemmed_results[:20])

    return stemmed_results

# --- EXECUTION PIPELINE ---

# Step 0: Mount Google Drive
drive.mount('/content/gdrive')

# Step 1: Read Files
speech_2006_content = read_speech_file(file_2006)
speech_2005_content = read_speech_file(file_2005)

# Step 2: Process Data (if reading was successful)
if speech_2006_content:
    processed_2006 = process_text(speech_2006_content, "2006 Speech")

if speech_2005_content:
    processed_2005 = process_text(speech_2005_content, "2005 Speech")

# You can now use the 'processed_2006' and 'processed_2005' lists for further analysis
# (e.g., frequency counting, comparison, visualization).
```

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer

try:
    nltk.data.find('corpora/state_union')
except LookupError:
    nltk.download('state_union')

try:
    # Explicitly downloading 'averaged_perceptron_tagger_eng' as suggested by error
    nltk.data.find('taggers/averaged_perceptron_tagger_eng')
except LookupError:
    nltk.download('averaged_perceptron_tagger_eng')

train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")

custom_sent_tokenizer = PunktSentenceTokenizer(train_text)

tokenized = custom_sent_tokenizer.tokenize(sample_text)

def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            chunkGram = r"""Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?}"""
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            # chunked.draw() # Commented out as it opens a new window, which might not be desired in Colab
            for subtree in chunked.subtrees():
                print(subtree)

    except Exception as e:
        print(str(e))

process_content()
```

```python
# --- IMPORTS ---
from google.colab import drive
import os
import spacy # Import spaCy
import pandas as pd
from itertools import chain

# --- CONFIGURATION ---
# Load the small English spaCy model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    print("Loading spaCy model failed. Did you run: !python -m spacy download en_core_web_sm")
    exit()

BASE_PATH = '/content/gdrive/MyDrive/datasets'
file_2006 = '2006-GWBush.txt'
file_2005 = '2005-GWBush.txt'

# Stop words from spaCy's vocabulary
SPACY_STOPWORDS = nlp.Defaults.stop_words

# --- 1. FILE READING FUNCTION (No Change) ---
def read_speech_file(filename):
    """Reads a file from the defined Google Drive path."""
    filepath = os.path.join(BASE_PATH, filename)
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            print(f"--- SUCCESS: Reading {filename} ---")
            return f.read()
    except FileNotFoundError:
        print(f"*** ERROR: File not found at {filepath}. Please check your path/spelling. ***")
        return None

# --- 2. NLP PROCESSING FUNCTION (UPDATED) ---
def process_text_spacy(text, file_year):
    """
    Performs spaCy processing: tokenization, POS, NER, and filtering.
    Returns a list of tuples: (token, pos_tag, ner_label, year)
    """
    if not text:
        return []

    print(f"\n--- PROCESSING: {file_year} Speech ---")

    # Process the text in one pass using spaCy
    doc = nlp(text)

    processed_tokens = []

    for token in doc:
        # Filter: Skip tokens that are stopwords, punctuation, or just whitespace
        if token.is_stop or token.is_punct or token.is_space or not token.text.strip():
            continue

        # Compile data for each token
        token_data = (
            token.text,                 # The actual word/token
            token.pos_,                 # Part-of-Speech Tag (e.g., NOUN, VERB)
            token.ent_type_,            # Named Entity Label (e.g., ORG, GPE, DATE)
            file_year                   # The year column
        )
        processed_tokens.append(token_data)

    print(f"Total tokens after filtering: {len(processed_tokens)}")
    return processed_tokens

# --- EXECUTION PIPELINE ---
drive.mount('/content/gdrive')

# Step 1: Read Files and Store Content
speech_2006_content = read_speech_file(file_2006)
speech_2005_content = read_speech_file(file_2005)

# Step 2: Process Data and Store Structured Lists
# We pass the content AND the year (2006 or 2005)
processed_2006_list = process_text_spacy(speech_2006_content, 2006)
processed_2005_list = process_text_spacy(speech_2005_content, 2005)

# --- 3. MERGE AND CREATE DATAFRAME ---
print("\n--- 3. Merging Data into DataFrame ---")

# Concatenate the two lists
merged_data = list(chain(processed_2006_list, processed_2005_list))

# Create a Pandas DataFrame with all four columns
df_nlp = pd.DataFrame(merged_data, columns=['Token', 'POS', 'NER_Label', 'Year'])

# Display the results
print("\n--- Final Merged DataFrame Head ---")
print(df_nlp.head(100))

print("\n--- DataFrame Info ---")
print(df_nlp.info())
```

```python
import pandas as pd

# Define the entity types we are interested in
entity_types = ['GPE', 'ORG', 'PERSON', 'DATE', 'VB']

print("\n--- NER Frequency Analysis (Top 5 Entities) ---")

# Filter for tokens that have a recognized NER label and are one of the types we want
ner_data = df_nlp[df_nlp['NER_Label'].isin(entity_types)]

# Group by Year and Entity Label, then count the tokens (words)
ner_counts = ner_data.groupby(['Year', 'NER_Label', 'Token']).size().reset_index(name='Count')

for year in ner_counts['Year'].unique():
    print(f"\n***** Year: {year} *****")

    for entity_type in entity_types:
        # Filter the counts for the current year and entity type
        top_entities = ner_counts[
            (ner_counts['Year'] == year) &
            (ner_counts['NER_Label'] == entity_type)
        ]

        # Sort and display the top 5
        top_entities = top_entities.sort_values(by='Count', ascending=False).head(5)

        if not top_entities.empty:
            print(f"- Top 5 {entity_type} Mentions:")
            print(top_entities[['Token', 'Count']].to_string(index=False))
        else:
            print(f"- No {entity_type} entities found.")
```

```python
import pandas as pd
import numpy as np # Needed for NaN values
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# --- ASSUMPTION: df_nlp is your DataFrame with the 'Token' and 'POS' columns populated ---
if 'df_nlp' not in globals():
    print("Error: DataFrame 'df_nlp' not found. Please run the previous consolidated code block first!")
    exit()

print("--- Calculating Polarity Scores ONLY for Verbs (VADER) ---")

try:
    nltk.data.find('sentiment/vader_lexicon')
except LookupError:
    nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to get the VADER composite score only for Verbs
def get_verb_polarity(row):
    """
    Returns the VADER compound score if the token is a VERB, otherwise returns NaN.
    """
    if row['POS'] == 'VERB':
        # VADER analysis returns a dictionary of scores (neg, neu, pos, compound)
        scores = sia.polarity_scores(row['Token'])
        return scores['compound']
    else:
        # Assign NaN for non-verbs to clearly distinguish them from neutral verbs (score close to 0)
        return np.nan

# 1. Apply the function to the DataFrame row-wise to create the new column
# axis=1 ensures the function receives the entire row, allowing access to 'POS' and 'Token'
df_nlp['Verb_Polarity_Score'] = df_nlp.apply(get_verb_polarity, axis=1)

# 2. Add a descriptive column to categorize the verb-specific polarity
# Use the same thresholds but handle NaN values
def categorize_verb_sentiment(score):
    if pd.isna(score):
        return np.nan # Keep non-verbs as NaN
    elif score >= 0.05:
        return 'Positive Verb'
    elif score <= -0.05:
        return 'Negative Verb'
    else:
        return 'Neutral Verb'

df_nlp['Verb_Sentiment'] = df_nlp['Verb_Polarity_Score'].apply(categorize_verb_sentiment)

# 3. Display the Results with the new columns
print("\n--- DataFrame Head with Verb-Specific Sentiment Columns ---")
# Add 'POS_Descriptor' to the list of columns if it exists, otherwise omit it.
# This prevents an error if 'POS_Descriptor' was not created in previous steps.
columns_to_display = ['Token', 'Year', 'POS']
if 'POS_Descriptor' in df_nlp.columns:
    columns_to_display.append('POS_Descriptor')
columns_to_display.extend(['Verb_Polarity_Score', 'Verb_Sentiment'])
print(df_nlp[columns_to_display].head(20).replace(np.nan, ''))

print("\n--- Verb Sentiment Analysis Summary by Year ---")

# Group by year and count the verb-only sentiment distribution
verb_sentiment_summary = df_nlp.groupby('Year')['Verb_Sentiment'].value_counts(normalize=True).mul(100).unstack(fill_value=0)
print(verb_sentiment_summary.round(2))
```
